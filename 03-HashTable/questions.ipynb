```python
def count_word_frequency(words):
    my_dict = {}
    for i in words:
        my_dict[i] = my_dict.get(i, 0) + 1
    return my_dict

words = ['apple', 'orange', 'banana', 'apple', 'orange', 'apple'] 
count_word_frequency(words)
```

```python
def merge_dicts(d1, d2):
    d3 = d1.copy()
    for key, value in d2.items():
        d3[key] = d3.get(key, 0) + value
    return d3

dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 3, 'c': 4, 'd': 5}
merge_dicts(dict1, dict2)
```

```python
def max_value_key(d):
    return max(d, key=d.get) # Get returns value associated with each key
    

my_dict = {'a': 5, 'b': 9, 'c': 2}
max_value_key(my_dict) # b
```

```python
def reverse_dict(my_dict):
    d = {}
    for key, value in my_dict.items():
        d[my_dict[key]] = key
    return d

my_dict = {'a': 1, 'b': 2, 'c': 3}
reverse_dict(my_dict)
```

```python
def filter_dict(my_dict, condition):
    return {k: v for k, v in my_dict.items() if condition(k, v)}

my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4} 
filter_dict(my_dict, lambda k, v: v % 2 == 0) 
```

```python
def check_same_frequency(list1, list2):
    def count_elements(lst):
        counter = {}
        for element in lst:
            counter[element] = counter.get(element, 0) + 1
        return counter
    
    return count_elements(list1) == count_elements(list2)

list1 = [1, 2, 3, 2, 1]
list2 = [3, 1, 2, 1, 3]
check_same_frequency(list1, list2)
```
